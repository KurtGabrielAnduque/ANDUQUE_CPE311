We want to look at data for the Facebook, Apple, Amazon, Netflix, and Google (FAANG) stocks, but we were given each as a separate CSV file. Combine them into a single file and store the dataframe of the FAANG data as faang for the rest of the exercises: 

1. Read each file in. 
2. Add a column to each dataframe, called ticker, indicating the ticker symbol it is for (Apple's is AAPL, for example). This is how you look up a stock. Each file's name is also the ticker symbol, so be sure to capitalize it. 
3. Append them together into a single dataframe. 
4. Save the result in a CSV file called faang.csv.

```python
import pandas as pd
import numpy as np

# loading every csv
apple = pd.read_csv('aapl.csv')
amzn = pd.read_csv('amzn.csv')
fb = pd.read_csv('fb.csv')
goog = pd.read_csv('goog.csv')
nflx = pd.read_csv('nflx.csv')
```

```python
# Add a column to each dataframe, called ticker, indicating the ticker symbol it is for (Apple's is AAPL, for example). 
#This is how you look up a stock. Each file's name is also the ticker symbol, so be sure to capitalize it.

# In this part Before I merge all the dataframes I creat a column Ticker to Indicate what Company of every row
apple['Ticker'] = 'AAPL'
amzn['Ticker'] = 'AMZN'
fb['Ticker'] = 'FB'
goog['Ticker'] = 'GOOG'
nflx['Ticker'] = 'NFLX'
```

```python
print(apple.shape)
print(amzn.shape)
print(fb.shape)
print(goog.shape)
print(nflx.shape)
# here I check the shape to calculate the expected row I should have after merging
# since all of them are 251 i should have 1255 rows after merging
```

```python
# Append them together into a single dataframe.
merge_df = apple.merge(amzn, how = 'outer')
# here I first merge the apple and amzn first I just use the outer since all of the columns are the same
```

```python
merge_df = merge_df.merge(fb, how = 'outer')
# now the merge_df is a dataframe that has a combination of Apple and amazon Im just going to use  this df to merge with the
# rest of the company
```

```python
merge_df = merge_df.merge(goog, how = 'outer')
```

```python
merge_df = merge_df.merge(nflx, how = 'outer')
```

```python
# here I check If there are nulls after the merging
merge_df.isnull().sum()
```

```python
# check the first 5 rows of the merge_df
merge_df.head()
```

```python
# check the shape if we successfully able to reach our expected number of row using shape function
merge_df.shape
```

```python
merge_df.to_csv('merge_dataframe.csv',index = False)
# Finally I will save the merge_df into the csv
```

- With faang, use type conversion to change the data column into a datetime and volume column into integers. Then, sort by date and ticker.

```python
# use type conversion to change the date column into a datetime and volume column into integers.

# here I change the date column into datetime datatype
# next is I change the datatype of column volumn from float into integer
merge_df['date'] = pd.to_datetime(merge_df['date'])
merge_df['volume'] = merge_df.volume.astype('int64')
```

```python
merge_df.dtypes
```

```python
# sort by date and ticker.
# Here I arrange the dataframe in ascending format bast on date and Ticker
merge_df.sort_values(['date','Ticker'], ascending = [True,True])
```

```python
# Find the sevens rows with the highest values for volume.
# Here I find the Top 7 rows with the Highest volume using the nlargest function
merge_df.nlargest(7,'volume')
```

```python
#  Use metl() to make it comepletely long format. Hint: data and ticker are our ID variable (they - uniquely identify each row). 
# We need to melt the rest so that we don't have separate columns for open, high, low, close, and volume.

merge_df = merge_df.melt(
    id_vars = ['date','Ticker'],
    var_name = 'variable_value',
    value_name = 'value'
)
```

- Using web scraping, search for the list of the hospitals, their address and contact information. Save the list in a new csv file, hospitals.csv.

```python
# Inorder to execute the webscraping
# I first download the bs4 for analyzing the html
# Next is I download the requests to get the html code of the desired website

from bs4 import BeautifulSoup
import requests

source = requests.get('https://valuecarehealth.com/search-by-hospital/').text

soup = BeautifulSoup(source,'html')
# If Im going to write the soup.prettify it would be very long since it is the whole html
```

```python
header = soup.find('thead')
# here since im only focused on getting the Table header 
# here I just find the thead to get column names and insert it into a dataframe
```

```python
header.prettify()
#here we can observer that it is still composed of html format I need to use the strip inorder to get the desired informations
```

```python
# here stip every th to get the desire information of every column where I avoid getting html codes
head = [title.text.strip() for title in header.find_all('th')][0:3]
print(head)
```

```python
import pandas as pd
# here I start importing the pandas to insert the column that i get recently into my dataframe
df1 = pd.DataFrame(columns = head)
```

```python
df1
```

```python
# Now for the data of every row part
# this can be found in the 'tr' part of the website if you are going to inspect it
# so here I just find_all the tr

table = soup.find_all('tr')

#Since the every tr is composed of  many table data cell 
# During t he iteration of the every tr
# I also find all the td of every tr
# then strip all of it to get the Hospital Name, Address, Contact Number
# there are some td that is empty so I have to check that if the len of the datas is equal or greater than 3 I will insert it into my dataframe
for row in table[1:]:
    rows = row.find_all('td')
    datas = [data.text.strip() for data in rows][0:3]
    if len(datas) == 3:
            length = len(df1)
            df1.loc[length] = datas
```
