Using the meteorite data from the Meteorite_Landings.csv file, create a pivot table that shows both the number of meteorites and the 95th percentile of meteorite mass for those that were found versus observed falling per year from 2005 through 2009 (inclusive). Hint: Be sure to convert the year column to a number as we did in the previous exercise.

```python
import pandas as pd
import numpy as np

df = pd.read_csv('Meteorite_Landings.csv')
#load the dataset meteorite
```

```python
df
```

```python
# Using the meteorite data from the Meteorite_Landings.csv file, 
# create a pivot table that shows both the number of meteorites and the 95th percentile of meteorite mass for those that were found versus 
# observed falling per year from n2005 through 2009 (inclusive). 
# Hint: Be sure to convert the year column to a number as we did in the previous exercise.

'''in this part I try to get the year in the column year of the dataset my approach is to use the split of strings then get the desired list'''
df['year'] = df['year'].str.split().str[0]
df['year'] = df['year'].str.split('/').str[2]
```

```python
df
# load the data to check if I was able to take the year in the column year
```

```python
df.isnull().sum()
# now is the phase of data cleaning 
# in this case i need to check if there are nan values in my column year
# now there are 291 row with empty year
```

```python
df.dropna(subset = ['year'], inplace = True)
# Here I remove the rows with empty year to avoid errors when im doing the analysis
# since the number of rows that are empty is below 5% of the whole dataset i just remove them
```

```python
df.reset_index(drop = True)
# after removing the row with empty year I rearrange the index
```

```python
df.dtypes
# here I check the datatypes of my dataframe
# I observe that my year is not in numerical
```

```python
df['year'] = df.year.astype('int64')
# here I need to turn the year datatype into numerical because I will use it for creating conditions
```

```python
df.dtypes
# now I check if I have change the datatype of column year successfully
```

```python
# 2005 through 2009
filtered_dataframe = df[(df['year'] >= 2005) & (df['year'] <= 2009)]
# here the task is to create an observation in the dataset that ranges from year 2005 and 2009
```

```python
filtered_dataframe['year'].unique()
# Check the list of unique values of the column year to see if we successfully get the desired year
```

```python
filtered_dataframe.sort_values(['year'],ascending = True)
# here I rearrange my dataframe based on the year column in ascending order
```

```python
# here I created pivot table 
# my index will be the year
# my columns will be the count and mass of the 95th percentile of each year
# inorder to do that I use to aggfunc count to the number of fell and found per year
# I use quantile(0.95) to get the 95th percentile
filtered_dataframe.pivot_table(index="year",columns="fall",values="mass (g)",aggfunc=["count", lambda massg: massg.quantile(0.95)])
```

```python
# Using the meteorite data from the Meteorite_Landings.csv file, compare summary statistics of the mass column for the meteorites that were found versus observed falling.
df.groupby(['fall'])['mass (g)'].describe()
#here I use groupby function to create a statiscal analysis of the mass for every meteorites that found and the once that only observe falling.
```

1. Using the taxi trip data in the 2019_Yellow_Taxi_Trip_Data.csv file, resample the data to an hourly frequency based on the dropoff time. Calculate the total trip_distance, fare_amount, tolls_amount, and tip_amount, then find the 5 hours with the most tips.

```python
import pandas as pd

df2 = pd.read_csv('2019_Yellow_Taxi_Trip_Data.csv',parse_dates = ["tpep_dropoff_datetime"])
# in this part I already change the datatype of column tpep_dropoff_datetime into datetime datatype since we are going to use it later
```

```python
df2.head()
#load the dataframe first five
```

```python
df2.dtypes
# here I check if I have change the datatype of dropoff column successfully
# and base on the output it turned into datetime datatype
```

```python
df2.set_index("tpep_dropoff_datetime", inplace=True)
# here I set the dropoff column as my index
# since we are going to resample it my hour later
```

```python
df2.head()
# here I check if I have the index is now the dropoff datetime
```

```python
# resample the data to an hourly frequency based on the dropoff time. 
hourly_data = df2.resample("h").sum()
# now we can resmple it by hour
```

```python
hourly_data.head()
# here as we load the dataframe
# we can observe that every data are now calculated per hour
```

```python
# Calculate the total trip_distance, fare_amount, tolls_amount, and tip_amount,
hourly_data[['trip_distance','fare_amount','tolls_amount','tip_amount']]
# here we check the per hour of the needed column
```
